Competitive Advantage
- AI improves network effects and innovation through new ways of applying data 
    - Porter's Five Forces and how to maximize advantage; 
        - Industry rivalry
            - Intensity of rivalry between firms in industry; intensity lowers profitability.
        - Threat of new entrants
            - Barriers to entry lower threat
        - Threat of subsitutes
            - More substitutions increases competitive pressure; Can happen very quickly (streaming, Uber, etc.)
        - Bargaining power of buyers
            - Powerful buyers can lower prices or increase COGS
        - Bargaining power of suppliers
            - Powerful suppliers can lower margins or increase prices
    - Network effects; increasing value of product when more people use it
    - Value capture dynamics; How to retain and monetize value 
- Modern competitive advantage is based on shaping, controlling, and harvesting networks.
    - Implication; Organizations that can aggregate data flowing between businesses and extract value through analytics are gaining competitive advantages.
    - In the modern market, what gives firms competitive advantages?
        - Cost, logistics, continuous innovations

Value Chain and Network Effects
- How to generate/capture value from users on a platform
- Network effects;
    - More users = more reasons to use social aspect of platform
    - More users = more developers/suppliers in the 'platform ecosystem' - improved suppliers
    - More users = more data and exponential analytical potential
    - More users = more value for external firms who sell on the platform - free work and platform stickiness
- Where does value go (*appropriability of value*)?
    - Examples;
        - Apple collecting large commissions on app sales
        - Google charging advertisers for posting ads
        - Uber taking percentages of rideshare costs from drivers
    - Factors;
        - Control over key assets (data, algorithm, user base)
        - Switching costs (trying to leave Apple ecosystem)
        - Government regulation and bargaining power of suppliers
- Multihoming; Using multiple platforms at once (as buyer or supplier)
    - Uber/Lyft, Netflix/HBO, Youtube/TikTok, Amazon/eBay
    - Easier multihoming = Fiercer competition

Disintermediation and Value Capture
- When platforms (and other businesses) cut out middlemen to connect directly to consumers
    - Expedia disintermediates travel agents
    - Spotify disintermediates record stores/distributors 
    - Glossier disintermediates brick and mortar 
- Effects;
    - Lowers control incumbent has over key assets
    - Lowers costs of switching due to less intermediaries
    - Regulators can force disintermediation 

Network Bridging and Value Capture
- Connecting networks to try and generate value out of the connection
    - LinkedIn bridges professional groups and separate HR systems
    - Amazon Marketplace bridges beetween multiple suppliers/industries
    - App Store bridges developers and users directly
