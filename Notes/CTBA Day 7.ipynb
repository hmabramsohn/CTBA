**The Analytical Problem**
- Gen AI is running into diminishing returns/no returns at all; skepticism about gen AI is spreading
- Data as oil;
    - Data is not scarce, but it does need to be refined to be valuable
- When done right, digital transformation can disrupt industries on fundamental levels
    - Consumer expectations and behavior change irrevocably as well
    - Expectations are high; personalized, instant, and consistent service 

*Analytics in Business*
- Analytics can drive internal efficiencies and theories, but often the value is not delivered in practice or to the customer
- To deliver real value, analytical tools need to be simplified, dispersed, and adopted across the company in general; not just for the analytical team

*Augmented Analytics*
- Intuitive tools that provide insight/information to a broad swathe of employees using a foundation of mixed, quality data
    - Excel and even PowerBI are too niche and require too much expertise to be useful at large scale
- Moving corporations towards a culture of referring to real data before making decisions; insight over intuition
    - Cultural norms around analysis, curiosity, and communication
- Analytical Maturity framework;
    - Data Reactive
    - Data Active - COEs, analytics strategy, foundational technology
    - Data Progressive - Focus on quality, operationalized machine learning, analytics in workflows
    - Data Fluent - Analytics embedded in culture and workflows
- Reliant on IPTOP framework;
    - Infrastructure
    - People
    - Tools
    - Organization - Centralizing or decentralizing data 
    - Processes
- SPEC framework;
    - Strategy - Align analytics vision with business objectives
    - People & Organization - Build data talent, combine business and analytical roles
    - Ecosystem - Reliable, secure, accessible data infrastructure
    - Culture - Collaboration, experimentation, and data-driven mindsets 

**Augmented Analytics**
- Finding the sweet spot between implementing automation technology and retaining human talent and judgement
    - Breaking down the divide between data experts and businesspeople; analytics as a core part of organizational workflows
- Five key components framework;
    - People
    - Technology
    - Analytical Leverage
    - Business Task
    - Better Way
- Five traits of effective AA framework;
    - Insightful
    - Integrated into workflows
    - Invisible/seamless
    - Indispensable - consistent
    - Inclusive across all workers

*Bias framework*
- Different types of bias are easier for humans or analytics to mitigate
    - Survivorship bias - Analytics
    - Historical bias - Humans
    - Cognitive bias - Both
    - Availability bias (Focusing on available data even if misrepresentative) - neither

*AI Archetypes framework*
- Tabular data; 
    - Supervised ML
    - Collecting training data; defining the features, training on data, iterating/refining model
- Nontabular data;
    - NLP
    - Audio and speech
    - Computer vision
- Multimodal data;
    - Gen AI
