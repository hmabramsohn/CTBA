**Rethinking the Firm**

Review
- Network effects and learning effects, which create additional value on top of physical products

*AI*
- Broad category, larger than genAI
- AI is not a bolt-on tool; it should be a strategic driver and shaped to solve specific business problems
- This is a transformation of how a business operates (culture shift) - transformative change
    - Mission, data architecture, governance, and product-centric agility
    - Reshaping of boundaries, industry position

Transformation
- Shifting from human-centric to algorithm-centric workflows; new roles, processes, and governance
- Data-driven thinking, agility, continuous experimentation; Focus on learning over intuition
- Led by leadership, based on trust and transparency 

AI Readiness
- Ability for an organization to adopt, implement, and scale AI
    - Digital infrastructure
        - Scalable, cloud-based, modern
    - Data maturity
        - Integrated data out of siloes into centralized platforms
        - Strong governance
    - Talent & Culture
        - Empowered cross-functional teams
    - Commitment from Leadership
        - Technical, strategic, and governance leaders
    - Experimentation & Learning Mindset
        - Agile iteration and modular development
- Alignment of firm strategy with digital and human resources

Qualitative research has shown that there may be a positive correlation between AI maturity and financial performance.

*Operating Models and AI*
- AI firms still need a consistent operating model; But built around data, algorithms, and digital platforms, instead of physical assets and labor. 

Traditional
- Physical supply-chains
- Human decision-making
- Marginal returns

AI
- Action stemming from learning derived from algorithms applied to data
- Operations are embedded into the digital platforms
- Growth can come from user interactions and automation, increasing marginal return 

Scale Economy
- Increased production = cost advantage; AI automation also contributes
- More operations, minimal cost increase

Scope Economy
- Effeciency formed by variety - offerings of multiple products/services
- AI helps leverage shared data and infrastructure to increase variety 

Alignment
- Creating value through cost, differentiation, etc.
- Capturing value through pricing, advertising, etc.
- Scaling operating model through increased volume and complexity, assisted by algorithms
- Scope increased by adding variety to offerings
- Learning to generate improvements and innovations

Continuous Learning Model
- Feedback loop of applying results from testing back to product design and decision-making
    - A/B testing on products, pricing, interfaces, etc.
    - Controlled trials

Strategic Implications
- Experimentation allows firms to adapt quickly 
- Agility is a competitive advantage; pivoting is necessary to adapt to changes and discoveries
- Strategy must be informed by real-world data and results instead of intuition and assumptions
- Scalable learning loops - feed experimentation back into business models 
- Stick to small experiments to lower risk of mistakes, which are inevitable

Human Bottlenecks 
- AI can enable higher complexity work by lowering the amount of time and mistakes caused by human decision-making when applied to low-level decisions
- Successfully automating low-level work provides a competitive advantage 

Irresistible Digital Bicycle
- AI is a supplement to human cognitive work; Humans can process and act on much larger amounts of data in a shorter amount of time, enabling new business 

*Tacit vs Strategic AI*

- AI can be used tactically to solve specific, short-term goals like operational improvements; a tool to automate a specific workflow
- Strategic AI reshapes an organization's long-term essence, rethinking value and business models 

*High Performers*

- Good understanding of operating model
- Commitment to talent acquisition and development for ML/AI professionals
- Good metrics and transparency
- Accessible, clean data 

*Toughest businesses for AI*

- Healthcare; 
    - Complexity of data; fragmented, categorical, unstructured, sensitive
    - High-stakes and reputational risk
    - Regulation and oversight
    - Low trust from stakeholders in the industry
    - Deep integration and investment required before useful strategically
